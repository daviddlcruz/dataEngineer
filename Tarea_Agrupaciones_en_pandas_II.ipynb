```python
import pandas as pd
import requests

url = 'https://raw.githubusercontent.com/ChrisnoSan/datasets/main/coordenadas.tsv'
respuesta = requests.get(url)
data = respuesta.text

data_limpia = data.replace("\t ", "\t")

file_path = "data_coordenadas.tsv"
with open(file_path, "w", encoding="utf-8") as f:
    f.write(data_limpia)

# Read the cleaned file using pandas
df_coordenadas = pd.read_csv(file_path, delimiter="\t")
```

```python
df_coordenadas.head(10)
```

```python
df_coordenadas.replace(["", None, pd.NA], "-")
```

```python
df_max_min = df_coordenadas.groupby('FUENTE').agg({
    'CODIGO': ['min', 'max'],  # Get min and max of CODIGO
    'DPTO': 'count'  # Count unique values in DPTO
})

df_max_min.head(40)
```
